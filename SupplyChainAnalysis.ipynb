```python
#shipping and delivery performance
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(data['Days for shipping (real)'], bins=30, kde=True, color="skyblue")
plt.title("Distribution of Actual Shipping Days")
plt.xlabel("Days for Shipping (real)")
plt.ylabel("Number of Orders")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Shipping Mode", y="Days for shipping (real)", data=data, hue="Shipping Mode", legend=False)
plt.title("Actual Shipping Days by Shipping Mode")
plt.xlabel("Shipping Mode")
plt.ylabel("Days for Shipping (real)")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x="Shipping Mode", y="Days for shipment (scheduled)", data=data, hue="Shipping Mode", legend=False)
plt.title("Scheduled Shipping Days by Shipping Mode")
plt.xlabel("Shipping Mode")
plt.ylabel("Days for Shipment (scheduled)")
plt.show()

```

```python
delivery_status = pd.crosstab(data['Shipping Mode'], data['Delivery Status'])

delivery_status.plot(kind='bar', stacked=True, figsize=(8,5), colormap="Paired")
plt.title("Delivery Status by Shipping Mode")
plt.xlabel("Shipping Mode")
plt.ylabel("Number of Orders")
plt.legend(title="Delivery Status")
plt.show()
```

```python
late_risk = data.groupby('Order Region')['Late_delivery_risk'].mean().sort_values()

plt.figure(figsize=(8,5))
sns.barplot(
    x=late_risk.index, 
    y=late_risk.values, 
    hue=late_risk.index,       # assign hue to avoid FutureWarning
    legend=False,              # hide duplicate legend
    palette="coolwarm"             # same palette style
)
plt.title("Average Late Delivery Risk by Region")
plt.xlabel("Order Region")
plt.ylabel("Late Delivery Risk (%)")
plt.xticks(rotation=45)
plt.show()

```

```python
# sales and profitability
top_products = data.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(
    x=top_products.values, 
    y=top_products.index, 
    hue=top_products.index,    # fix palette warning
    legend=False,
    palette="Set2"
)
plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="Sales", 
    y="Order Profit Per Order", 
    data=data, 
    hue="Order Profit Per Order",  # add hue for palette control if needed
    palette="coolwarm",
    legend=False,
    alpha=0.6
)
plt.title("Profit vs Sales per Order")
plt.xlabel("Sales")
plt.ylabel("Profit per Order")
plt.show()
```

```python
category_sales_profit = data.groupby("Category Name")[["Sales","Order Profit Per Order"]].sum().reset_index()

# plot Sales
plt.figure(figsize=(10,6))
sns.barplot(
    x="Category Name", 
    y="Sales", 
    data=category_sales_profit, 
    hue="Category Name", 
    legend=False, 
    palette="Set2"
)
plt.title("Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()

# plot Profit
plt.figure(figsize=(10,6))
sns.barplot(
    x="Category Name", 
    y="Order Profit Per Order", 
    data=category_sales_profit, 
    hue="Category Name", 
    legend=False, 
    palette="Set3"
)
plt.title("Profit by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="Order Item Discount", 
    y="Order Item Profit Ratio", 
    data=data, 
    hue="Order Item Profit Ratio", 
    palette="viridis", 
    legend=False, 
    alpha=0.5
)
plt.title("Discount vs Profit Ratio")
plt.xlabel("Order Item Discount")
plt.ylabel("Profit Ratio")
plt.show()

```

```python
delivery_market = pd.crosstab(data["Market"], data["Delivery Status"], normalize="index") * 100

delivery_market.plot(
    kind="bar", 
    stacked=True, 
    figsize=(10,6), 
    colormap="Paired"
)
plt.title("Delivery Status % by Market")
plt.xlabel("Market")
plt.ylabel("Percentage of Orders")
plt.legend(title="Delivery Status")
plt.show()
```

```python
# convert order date
data["order date (DateOrders)"] = pd.to_datetime(data["order date (DateOrders)"], errors="coerce")

monthly_sales = data.groupby([data["order date (DateOrders)"].dt.to_period("M"), "Market"])["Sales"].sum().reset_index()
monthly_sales["order date (DateOrders)"] = monthly_sales["order date (DateOrders)"].astype(str)

plt.figure(figsize=(12,6))
sns.lineplot(
    x="order date (DateOrders)", 
    y="Sales", 
    data=monthly_sales, 
    hue="Market", 
    palette="Set1"
)
plt.title("Monthly Sales Trend by Market")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
#Product Analysis
top_products_qty = data.groupby("Product Name")["Order Item Quantity"].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(
    x="Order Item Quantity", 
    y="Product Name", 
    data=top_products_qty, 
    hue="Product Name", 
    legend=False, 
    palette="Set2"
)
plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Product Name")
plt.show()
```

```python
least_profitable = data.groupby("Product Name")["Order Profit Per Order"].sum().sort_values().head(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(
    x="Order Profit Per Order", 
    y="Product Name", 
    data=least_profitable, 
    hue="Product Name", 
    legend=False, 
    palette="Set3"
)
plt.title("Top 10 Least Profitable Products")
plt.xlabel("Total Profit")
plt.ylabel("Product Name")
plt.show()
```

```python
category_sales = data.groupby("Category Name")["Sales"].sum().reset_index()

plt.figure(figsize=(8,8))
colors = sns.color_palette("Set2", len(category_sales))

plt.pie(
    category_sales["Sales"], 
    labels=category_sales["Category Name"], 
    autopct='%1.1f%%', 
    colors=colors, 
    startangle=90, 
    wedgeprops={'width':0.4}   # makes it a donut
)
plt.title("Category-wise Sales Contribution")
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="Order Item Discount", 
    y="Sales", 
    data=data, 
    hue="Order Item Discount", 
    palette="viridis", 
    legend=False, 
    alpha=0.5
)
plt.title("Discount vs Sales")
plt.xlabel("Discount Amount")
plt.ylabel("Sales")
plt.show()
```

```python
#Advance Analysis
plt.figure(figsize=(8,6))
corr = data[["Sales", "Order Profit Per Order", "Order Item Discount", "Order Item Quantity", "Product Price"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Key Metrics")
plt.show()
```

```python
product_sales = data.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).reset_index()
product_sales["Cumulative_Sales"] = product_sales["Sales"].cumsum()
product_sales["Cumulative_Percentage"] = 100 * product_sales["Cumulative_Sales"] / product_sales["Sales"].sum()

plt.figure(figsize=(12,6))
sns.barplot(
    x="Product Name", y="Sales", 
    data=product_sales.head(20), 
    hue="Product Name", legend=False, 
    palette="Set2"
)
plt.plot(product_sales["Product Name"].head(20), product_sales["Cumulative_Percentage"].head(20), color="red", marker="o")
plt.axhline(80, color="grey", linestyle="--")
plt.xticks(rotation=90)
plt.title("Pareto Analysis of Top 20 Products (80/20 Rule)")
plt.ylabel("Sales")
plt.show()
```

```python
# Convert to datetime if not already
data['order date (DateOrders)'] = pd.to_datetime(data['order date (DateOrders)'])

sales_trend = data.groupby(data['order date (DateOrders)'].dt.to_period("M"))["Sales"].sum().reset_index()
sales_trend['order date (DateOrders)'] = sales_trend['order date (DateOrders)'].astype(str)

plt.figure(figsize=(12,6))
sns.lineplot(
    x="order date (DateOrders)", y="Sales", 
    data=sales_trend, 
    marker="o", color="blue"
)
plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()
```

```python
shipping_profit = data.groupby("Shipping Mode")["Order Profit Per Order"].mean().reset_index()

plt.figure(figsize=(8,6))
sns.barplot(
    x="Shipping Mode", y="Order Profit Per Order", 
    data=shipping_profit, 
    hue="Shipping Mode", legend=False, 
    palette="Set3"
)
plt.title("Average Profit per Order by Shipping Mode")
plt.ylabel("Average Profit")
plt.show()
```

```python
clv = data.groupby("Customer Id").agg({
    "Sales": "sum",
    "Order Id": "nunique"
}).reset_index()

clv["Avg_Order_Value"] = clv["Sales"] / clv["Order Id"]

plt.figure(figsize=(10,6))
sns.histplot(clv["Avg_Order_Value"], bins=30, kde=True, color="teal")
plt.title("Distribution of Average Order Value per Customer")
plt.xlabel("Avg Order Value")
plt.ylabel("Number of Customers")
plt.show()
```
